```python
import torch
import matplotlib.pyplot as plt
```

```python
model_metrics = torch.load("best_model.pth", map_location=torch.device('cpu'))
epoch_list = list(range(93))
train_loss_list = model_metrics["train_loss_list"]
val_loss_list = model_metrics["val_loss_list"]
acc_list = model_metrics["acc_list"]
mae_list = model_metrics["mae_list"]

model_metrics["acc"]

```

```python
len(acc_list), len(epoch_list)
```

```python
def plot_single_metric(y_list, x_list, title, x_label, y_label):
    plt.title(title)
    plt.plot(x_list, y_list)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.grid(True)
    plt.show()
    
plot_single_metric(acc_list, epoch_list, "Classification Accuracy", "Epochs", "Accuracy")
```

```python
plot_single_metric(mae_list, epoch_list, "Regression Error", "Epochs", "MAE")
```

```python
plt.title("Loss over epochs")
plt.plot(epoch_list, train_loss_list, color="r", label="Train Loss")
plt.plot(epoch_list, val_loss_list, color="b", label="Val Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.legend()
plt.grid(True)
plt.show()
```
